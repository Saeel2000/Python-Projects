Now we will change index replace with rank based on max. transcation

```python
a['Amount.Of.Transaction'].replace(",","",inplace=True, regex=True)

a['Amount.Of.Transaction']=a['Amount.Of.Transaction'].astype('float')
a['Rank']=a['Amount.Of.Transaction'].rank(ascending=False,method='first')
a.sort_values('Rank')
```

```python
a.set_index('Rank',inplace=True)
```

```python
a
```

```python
plt.figure(figsize=(10,10))
aa=sns.countplot(x='Month',data=a)
plt.title("Counts of months")
for bars in aa.containers:
    aa.bar_label(bars)
```

```python
plt.figure(figsize=(6,5))
month_Transactions=a.groupby(['Year'],as_index=True)['No.Of.Banks.Using.Upi'].sum().sort_values(ascending=False)
ab=sns.barplot(data=a.head(15),x='Year',y='No.Of.Banks.Using.Upi')
plt.title('Year by NO. of banks')
for bars in ab.containers:
    ab.bar_label(bars)
```

```python
month_Transactions
```

```python
Year_column = 'Year'
Transaction_column = 'Amount.Of.Transaction.In.Bn'

grouped_a = a.groupby(Year_column)[Transaction_column].sum().reset_index()


plt.figure(figsize=(12,6))
plt.pie(grouped_a[Transaction_column], labels=grouped_a[Year_column], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors, textprops={'fontsize':5})
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Distribution of Transactions by Year')


# Add legends
plt.legend(title='Year', loc='upper right', bbox_to_anchor=(1, 0, 0.5, 1))

# Display the pie chart
plt.show()
```

```python
plt.figure(figsize=(20,10))
month_Transactions=a.groupby(['Month-Year'],as_index=True)['Volume.Of.Transaction.in.Mn'].sum().sort_values(ascending=False)
ax=sns.barplot(data=a.head(15),x='Month-Year',y='Volume.Of.Transaction.in.Mn')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Top15 vol. of Transaction with Month-year')
```
