# Plot a histogram of salaries

```python
plt.figure(figsize=(10, 6))
plt.hist(a['Salary'],edgecolor='black')

plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
```

# Distribution of salaries by country by using pie chart

```python
category_column = 'Country'
salary_column = 'Salary'

grouped_a = a.groupby(category_column)[salary_column].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.pie(grouped_a[salary_column], labels=grouped_a[category_column], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Distribution of Salaries by Category')

# Add legends
plt.legend(title='Categories', loc='upper right', bbox_to_anchor=(1, 0, 0.5, 1))

# Display the pie chart
plt.show()
```
